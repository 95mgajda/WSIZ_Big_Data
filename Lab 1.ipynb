```python
import pandas as pd
import numpy as np

from google.colab import drive
from google.colab import data_table
%load_ext google.colab.data_table

drive.mount('/content/gdrive')

#Ścieżka do pliku csv
path = "/content/gdrive/MyDrive/Colab Notebooks/argenziano.csv"

dane = pd.read_csv(path)

dane = dane.drop(columns=["ID", "Unnamed: 0"], errors='ignore')
print("Dane wczytane! Shape: ", dane.shape)
dane.head()

```

```python
# ROZBICIE + OCZYSZCZANIE
dane_split = dane.iloc[:, 0].str.split(';', expand=True)
dane_split.columns = [
    'ARGNZO_ATYPOWA_SIATKA_BARWNIKOWA',
    'ARGNZO_ATYPOWY_WZOR_NACZYNIOWY',
    'ARGNZO_NIEBIESKO_BIALAWA_WOALKA',
    'ARGNZO_NIEREGULARNE_SMUGI',
    'ARGNZO_PSEUDOWYPUSTKI',
    'ARGNZO_SMUGI_PROMIENISTE',
    'ARGNZO_NIEREGULARNE_ZABARWIENIE',
    'ARGNZO_NIEREGULARNE_PLAMY_GLOBULE',
    'ARGNZO_OBSZARY_REGRESJI',
    'WYNIK'
]

# NAJWAŻNIEJSZE: konwersja z czyszczeniem błędów!
feature_cols = dane_split.columns[:-1]  # wszystko oprócz WYNIK
for col in feature_cols:
    # Zamień nie-numeryczne na 0, potem do int
    dane_split[col] = pd.to_numeric(dane_split[col], errors='coerce').fillna(0).astype(int)

print("OK, Shape po czyszczeniu:", dane_split.shape)
print("Kolumn:", list(dane_split.columns))
print("\nPierwsze 5 wierszy:")
dane_split.head()
```

```python
dane_clean = dane_split.copy()  # gotowe dane do LAB

print("=== ZADANIE 2: Analiza argenziano.csv ===")
print("1. NAZWY KOLUMN:")
print(list(dane_clean.columns))

print("\n2. LICZBA PRZYPADKÓW I ATRYBUTÓW:")
print(dane_clean.shape)

print("\n3. PIERWSZY WIERSZ, PIERWSZA KOLUMNA:")
print(dane_clean.iloc[0, 0])

target_col = 'WYNIK'
y = dane_clean[target_col]
X = dane_clean.drop(columns=[target_col])

print("\n4. ODDZIELENIE DANYCH:")
print("X shape (cechy):", X.shape)
print("Y shape (decyzje):", y.shape)

print("\n5. TYPY ATRYBUTÓW:")
print(dane_clean.dtypes)

print("\n6. UNIKALNE WARTOŚCI ATRYBUTÓW:")
for col in X.columns:
    unique_vals = sorted(X[col].unique())
    print(f"{col}: {unique_vals}")

print("\n7. ATRYBUT DECYZYJNY 'WYNIK':")
print(y.value_counts())
print("Liczba konceptów:", y.nunique())
```

```python
print("=== ZADANIE 3/4: BRAKUJĄCE WARTOŚCI ===")

null_any = dane_clean.isnull().any()
null_sum = dane_clean.isnull().sum()
null_pct = (dane_clean.isnull().mean() * 100).round(2)

missing_df = pd.DataFrame({
    'Atrybut': dane_clean.columns,
    'Nulls (True/False)': null_any.values,
    'Ilość': null_sum.values,
    '%': null_pct.values
})

print(missing_df.to_string(index=False))
missing_df
```

```python
print("=== ZADANIE 5: UZUPEŁNIANIE BRAKÓW (symulacja) ===")

# Twój zbiór NIE MA brakujących, ale kod ogólny:
dane_imputed = dane_clean.copy()
target_col = 'WYNIK'

braki_przed = dane_imputed.isnull().sum().sum()
print(f"Braki przed: {braki_przed}")

# Kod z LAB (nie zmieni nic)
for col in dane_imputed.columns:
    if col == target_col:
        continue
    if dane_imputed[col].isnull().any():
        dane_imputed[col] = dane_imputed.groupby(target_col)[col]\
            .transform(lambda s: s.fillna(s.mode().iloc[0]))

braki_po = dane_imputed.isnull().sum().sum()
print(f"Braki po: {braki_po}")
print("Ok, zero zmian – dane są czyste")
```

```python
from sklearn.preprocessing import LabelEncoder

print("=== ZADANIE 6: KODOWANIE DANYCH ===")

# Dane już numeryczne (0,1,2) + target kategoryczny
X = dane_clean.drop(columns=['WYNIK'])
y_str = dane_clean['WYNIK']

# Koduj target: Nie_czerniak=0, Czerniak=1
le = LabelEncoder()
y = le.fit_transform(y_str)
print("Target zakodowany:", dict(zip(le.classes_, le.transform(le.classes_))))
print("y shape:", y.shape, "unikalne:", np.unique(y))

print("\nX (cechy) gotowe do modelowania:")
print("Shape:", X.shape)
print("Zakres wartości:\n", X.describe())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

print("=== ZADANIE 7: KLASYFIKATORY ===")

# Podział train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 3 modele
models = {
    "Drzewo decyzji": DecisionTreeClassifier(max_depth=5, random_state=42),
    "SVM": SVC(kernel="rbf", random_state=42),
    "kNN (k=5)": KNeighborsClassifier(n_neighbors=5)
}

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results.append({"Model": name, "Accuracy": acc})

    print(f"\n{name}:")
    print(f"Accuracy: {acc:.3f}")
    print(classification_report(y_test, y_pred,
                               target_names=le.classes_))

# Tabela porównawcza
results_df = pd.DataFrame(results).sort_values("Accuracy", ascending=False)
print("\n PORÓWNANIE MODELI:")
results_df
```
